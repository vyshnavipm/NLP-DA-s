```python
# Installing and importing necessary libraries and packages
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
import pandas as pd
```

```python
# Importing the Brown Corpus
from nltk.corpus import brown
```

```python
# Finding the categories
brown.categories()
```

```python
# Exploring the "news" category in the Brown corpus
temp =brown.words(categories='reviews')
```

```python
brown.raw(categories='reviews')
```

```python
# Readable version of the above raw text
news_sent = brown.sents(categories='reviews')
news_sent
```

```python
# Exploring the list of identifiers
nltk.corpus.brown.fileids()
```

```python
# Finding the tokens
from nltk import word_tokenize
tokens = word_tokenize(brown.raw())
tokens
```

```python
# Finding the size of the corpus (total number of tokens)
from nltk.probability import FreqDist
fdist = FreqDist(brown.words())
fdist.N()
```

```python
# Finding the number of tokens occurring in the corpus of each type
from nltk.probability import FreqDist
fdist = FreqDist(brown.words())
```

```python
fdist
```

```python
# Tagging each word
pos_tags = nltk.pos_tag(brown.words())
pos_tags
```

```python
# Counting different types of tokens/words
from collections import Counter
counts = Counter( tag for word,  tag in pos_tags)
print(counts)
```

```python
# Finding the number of words in the "government" category
fdist["government"]
```

```python
# Finding the frequency of the tokens and printing the three most occuring tokens
freq = nltk.FreqDist(brown.words())
print("Three most frequent tokens are ", freq.most_common(3))
```

```python
# Counting the total number of sentences in the corpus
print("The number of sentences =", len(brown.sents()))
```
